```python
# 1. Create new column: TotalScore = MathScore + EnglishScore
student_df['TotalScore'] = student_df['MathScore'] + student_df['EnglishScore']
print("\n1. Added 'TotalScore' column (MathScore + EnglishScore)")
print(student_df[['StudentID', 'Name', 'MathScore', 'EnglishScore', 'TotalScore']].head())
```

```python
# 2. Compute metrics
print("\n2. COMPUTED METRICS:")

# Average scores
avg_math = student_df['MathScore'].mean()
avg_english = student_df['EnglishScore'].mean()
print(f"• Average Math Score: {avg_math:.2f}")
print(f"• Average English Score: {avg_english:.2f}")
print(f"• Average Total Score: {student_df['TotalScore'].mean():.2f}")

# Highest and lowest scorers
highest_idx = student_df['TotalScore'].idxmax()
lowest_idx = student_df['TotalScore'].idxmin()

print(f"\n• Highest Scorer:")
print(f"  Name: {student_df.loc[highest_idx, 'Name']}")
print(f"  Math: {student_df.loc[highest_idx, 'MathScore']}, English: {student_df.loc[highest_idx, 'EnglishScore']}")
print(f"  Total Score: {student_df.loc[highest_idx, 'TotalScore']}")

print(f"\n• Lowest Scorer:")
print(f"  Name: {student_df.loc[lowest_idx, 'Name']}")
print(f"  Math: {student_df.loc[lowest_idx, 'MathScore']}, English: {student_df.loc[lowest_idx, 'EnglishScore']}")
print(f"  Total Score: {student_df.loc[lowest_idx, 'TotalScore']}")
```

```python

# 3. Display top and bottom students
print("\n3. TOP AND BOTTOM 10 STUDENTS:")

# Top 10 students
top_10 = student_df.nlargest(10, 'TotalScore')[['Name', 'MathScore', 'EnglishScore', 'TotalScore']]
top_10.index = range(1, 11)  # Renumber indices
print("\n• Top 10 Students:")
print(top_10.to_string())

# Bottom 10 students
bottom_10 = student_df.nsmallest(10, 'TotalScore')[['Name', 'MathScore', 'EnglishScore', 'TotalScore']]
bottom_10.index = range(1, 11)  # Renumber indices
print("\n• Bottom 10 Students:")
print(bottom_10.to_string())
```

```python

# 4. Plot visualizations
print("\n4. GENERATING VISUALIZATIONS...")

plt.figure(figsize=(15, 10))

# Histogram of Math scores
plt.subplot(2, 2, 1)
student_df['MathScore'].hist(bins=15, edgecolor='black', alpha=0.7, color='skyblue')
plt.title('Distribution of Math Scores', fontsize=14, fontweight='bold')
plt.xlabel('Math Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.axvline(avg_math, color='red', linestyle='--', linewidth=2, label=f'Mean: {avg_math:.1f}')
plt.legend()

# Bar chart of total scores for top 10 students
plt.subplot(2, 2, 2)
bars = plt.bar(range(len(top_10)), top_10['TotalScore'], color='lightgreen', edgecolor='black')
plt.title('Total Scores of Top 10 Students', fontsize=14, fontweight='bold')
plt.xlabel('Student Rank', fontsize=12)
plt.ylabel('Total Score', fontsize=12)
plt.xticks(range(len(top_10)), [f'#{i}' for i in range(1, 11)])
plt.ylim(120, 200)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}', ha='center', va='bottom', fontweight='bold')

# Pie chart of gender distribution
plt.subplot(2, 2, 3)
gender_counts = student_df['Gender'].value_counts()
colors = ['lightblue', 'lightpink']
plt.pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%', 
        colors=colors, startangle=90, explode=(0.05, 0), shadow=True)
plt.title('Gender Distribution', fontsize=14, fontweight='bold')

# Scatter plot of Math vs English scores
plt.subplot(2, 2, 4)
colors = ['blue' if g == 'Male' else 'red' for g in student_df['Gender']]
plt.scatter(student_df['MathScore'], student_df['EnglishScore'], c=colors, alpha=0.6, s=50)
plt.title('Math vs English Scores', fontsize=14, fontweight='bold')
plt.xlabel('Math Score', fontsize=12)
plt.ylabel('English Score', fontsize=12)
plt.grid(True, alpha=0.3)

# Add correlation coefficient
correlation = student_df['MathScore'].corr(student_df['EnglishScore'])
plt.text(30, 100, f'Correlation: {correlation:.3f}', fontsize=11, 
         bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8))

# Add legend for gender
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='blue', alpha=0.6, label='Male'),
                   Patch(facecolor='red', alpha=0.6, label='Female')]
plt.legend(handles=legend_elements, loc='lower right')

plt.tight_layout()
plt.savefig('section_c_student_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
print("✓ Student visualizations saved as 'section_c_student_analysis.png'")
```
